```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d =pd.read_csv('Power plant.csv')
```

```python
d
```

```python
#scatter plot using relplot
sns.relplot(x="V",y="AT",data=d)
```

```python
#hue semantic
sns.relplot(x="V",y="PE",hue="AT", data=d)
```

```python
t =sns.load_dataset("tips")
t
```

```python
sns.relplot(x="size",y="tip",data=t,kind="line")
```

```python
#bar plot
import numpy as np
sns.barplot(x="smoker",y="tip",data=t,estimator=np.std)
```

```python
# to show the histogram of tip
sns.displot(t['size'])
sns.displot(t['tip'])
sns.displot(t['smoker'])
sns.displot(t['day'])
```

```python
# to get lineplot
sns.lineplot(x="size",y="tip",data=t)
```
